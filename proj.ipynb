```python
# Dependencies and setup
%matplotlib ipympl
import pandas as pd
from pathlib import Path
```

```python
netflix_userbase_data_load = "../Resources/Netflix Userbase.csv"
```

```python
netflix_data_df = pd.read_csv(netflix_userbase_data_load)
```

```python
new_column_data_df = netflix_data_df[['Subscription Type','Country','Monthly Revenue','Age', 'Gender']]
```

```python
# Region Dataframes
NorthAmerica_data_df = new_column_data_df.loc[new_column_data_df["Country"].isin(["United States", "Canada", "Mexico"])]
SouthAmerica_data_df = new_column_data_df.loc[new_column_data_df["Country"] == "Brazil"]
EU_data_df = new_column_data_df.loc[new_column_data_df["Country"].isin(["Spain", "Germany", "France", "Italy", "United Kingdom"])]
Aus_data_df = new_column_data_df.loc[new_column_data_df["Country"] == "Australia"]

```

```python
netflix_data_binned = netflix_user_data.drop(columns= ['Monthly Revenue', 'Join Date',
                                                       'Last Payment Date', 
                                                       'Subscription Type','Plan Duration',
                                                       'User ID'])
```

```python
#binning- I want to see if the type of devices used matter by age group
# Create bins in which to place values based upon IMDB vote count
bins = [26,  #1
        29,  #2
        35,  #3
        40,  #4
        45,  #5
        51]  #6

# Create labels for these bins
group_labels = ["26-29",       #1
                "30-35",       #2 
                "36-40",       #3 
                "41-45",       #4 
                "46-51"]       #5
```

```python
#Place the data series into a new column inside of the DataFrame

netflix_data_binned["User_Age_Groups"] = pd.cut(netflix_data_binned["Age"], bins=bins, labels=group_labels)
netflix_data_binned
```

```python
devices_used_by_age = netflix_data_binned.groupby(["User_Age_Groups", "Device"])["Device"].count()
devices_used_by_age_df = pd.DataFrame(devices_used_by_age)


devices_used_by_age_df.rename(columns={'Device': 'Device Count'}, inplace=True)
devices_used_by_age_df
```

```python
#this data frame would be the easiest to map
devices_used_by_age_reset = devices_used_by_age_df.reset_index()
devices_used_by_age_reset
```
